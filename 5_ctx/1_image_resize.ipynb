```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
with open('./dataset/dataset_1.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
def resize_image(image,size,interpolation): #插值方法
    return cv2.resize(image, size, interpolation = interpolation)
```

```python
plt.figure(figsize=(12,12))
random_list = []
for i in range(1,17):
    index = np.random.randint(10000)
    random_list.append(index)
    plt.subplot(4,4,i)
    plt.imshow(data[index][0])
plt.show()
```

```python
plt.figure(figsize=(12,12))
for i in range(1,17):
    index = random_list[i-1]
    image = data[index][0]
    new_image = resize_image(image,(24,24),cv2.INTER_CUBIC)
    plt.subplot(4,4,i)
    plt.imshow(new_image)
plt.show()
```
